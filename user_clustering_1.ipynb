```python
from movielens import *
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.cluster import KMeans
import sys
import time
import pickle
user = []
item = []
rating = []
rating_test = []

# Load the movie lens dataset into arrays
d = Dataset()
d.load_users("data/u10.user", user)
d.load_items("data/u.item", item)
d.load_ratings("data/u10.base", rating)
d.load_ratings("data/u10.test", rating_test)

n_users = len(user)
n_items = len(item)
```

```python
print (rating)
```

```python
utility = np.zeros((n_users, n_items))
for r in rating:
    utility[r.user_id-1][r.item_id-1] = r.rating
```

```python
for i in range(0,n_items):
    print (utility[0][i])
```

```python
for i in range(n_users):
    rated = np.nonzero(utility[i]) #np.nonzero returns indices of the elements that are non-zero.
    n = len(rated[0])
    if n != 0:
        user[i].avg_r = np.mean(utility[i][rated])
    else:
        user[i].avg_r = 0.
```

```python
test = np.zeros((n_users, n_items))
for r in rating_test:
    test[r.user_id - 1][r.item_id - 1] = r.rating
```

```python
for i in range(0,n_items):
    print (test[0][i])
```

```python
cluster = KMeans(n_clusters=10)
cluster.fit_predict(utility)
```

```python
print (utility[0][19])
```

```python
user_id = 0
item_id = 19
```

```python
cluster_number = cluster.labels_[user_id]
print (cluster_number)
```

```python
c = []
for i in range(0,n_users):
  if(cluster.labels_[i] == cluster_number):
      c.append(i)
c.remove(user_id)
print (c)
```

```python
y = []
for user in c:
    x = utility[user][item_id]
    y.append(x)
    
print (y)
```
